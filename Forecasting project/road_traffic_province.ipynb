import library ที่ต้องการใช้ได้แก่ pandas และ geopandas สำหรับอ่านไฟล์ data

```python
import pandas as pd
import geopandas as gpd
```

อ่านไฟล์ข้อมูลการจราจรเฉลี่ยของแต่ละจังหวัด

```python
traffic = pd.read_csv(r'E:\WFH\traffic-province.csv')
traffic
```

แปลงข้อมูลเป็น list เพื่อนำไป filter

```python
traffic['Province'].tolist()
```

อ่านไฟล์ข้อมูลที่เป็น polygon ของทุกจังหวัด

```python
polygon_province = gpd.read_file(r'C:\Users\ihere\Desktop\Thailand_province.geojson')
polygon_province
```

ทำการ filter ข้อมูลเฉพาะจังหวัดที่เรามีข้อมูลการจราจรเฉลี่ย

```python
province_traffic = polygon_province[polygon_province['PROV_NAME'].isin(traffic['Province'].tolist())]
province_traffic = province_traffic.reset_index(drop=True)
province_traffic
```

เพิ่มคอลั่ม Traffic_avg ลงในข้อมูล polygon ของจังหวัดที่ข้อมูลการจราจรเฉลี่ย
จากนั้นเพิ่มค่าการจราจรเฉลี่ยจากข้อมูลการจราจร

```python
province_traffic.insert(11, 'Traffic_avg', 0.0)
for i in province_traffic.index:
    for j in traffic.index:
        if province_traffic['PROV_NAME'][i] == traffic['Province'][j]:
            province_traffic.at[i, 'Traffic_avg'] = traffic['Average'][j]
province_traffic
```

เขียนข้อมูล polygon ที่มีข้อมูลการจราจรเฉลี่ยออกมาเป็นไฟล์ geojson

```python
province_traffic.to_file("E:\WFH\province_traffic.geojson", driver='GeoJSON', encoding='utf-8')
```
